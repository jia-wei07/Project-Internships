```python
# import modules
from datetime import datetime
from pandas_datareader import data as pdr
import yfinance as yf
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Jia_W/OneDrive/Desktop/stock_code.csv')
df.shape
df2 = df.to_dict('dict')
df2
```

```python
for i in range(len(df2['stock_code'])):
    start = datetime(2018,1,1) 
    end = datetime(2023,1,15) 
    stock = yf.Ticker(df2['stock_code'][i]) 
    table = stock.history(start=start, end=end)
    
    frame=pd.DataFrame(table)
    frame['stock name']= df2['stock_company'][i]

    frame.to_csv('historical_data.csv', mode='a', index=True, header=False)
```
